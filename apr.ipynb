```python
from itertools import combinations

def apriori(dataset,min_support=0.5,min_conf=0.5):

    itemset=[]
    items=set()

    for transaction in dataset:
        for item in transaction:
            items.add(item)

    for i in range(1,len(items)+1):
        for combo in combinations(items,i):
            count = 0
            for transaction in dataset:
                if set(combo).issubset(transaction):
                    count += 1
            support=count/len(dataset)

            if support>=min_support:
                itemset.append((combo,round(support,2)))

    rules=[]
    for item,support in itemset:
        if len(item)>1:
            for i in range(1,len(item)):
                for left in combinations(item,i):
                    right=tuple(sorted(set(item)-set(left)))
                    left_count=0
                    for transaction in dataset:
                        if set(left).issubset(transaction):
                            left_count+=1
                    if left_count>0:
                        confidence=(support*len(dataset))/left_count
                        if confidence>min_conf:
                            rules.append((left,right,round(confidence,2)))

    print(f"Frequent ItemSets with support>= {min_support}")
    for combo,support in itemset:
        print(f"Itemset: {combo} | Support: {support}")

    print(f"\nAssociation Rules with Confidence >={min_conf}")
    for left,right,confidence in rules:
        print(f"Rule: {left}->{right} | Confidence: {confidence}")


transcations = [
        ['milk', 'bread', 'butter'],
    ['milk', 'bread'],
    ['milk', 'diaper', 'beer', 'bread'],
    ['bread', 'butter'],
    ['milk', 'diaper', 'bread', 'butter'],
    ['diaper', 'beer'],
    ['milk', 'bread', 'butter', 'beer'],
    ['bread', 'butter']
]

apriori(transcations, min_support=0.3, min_conf=0.6)



```
